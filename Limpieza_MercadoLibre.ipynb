```python
from google.colab import files
import pandas as pd
```

```python
uploaded = files.upload()

df = pd.read_csv("mercadolibre_laptops.csv")

df.head(100)
```

```python
#Agregando la columna tienda
df["Tienda"] = "Mercado Libre"
```

```python
df.isnull().sum()
```

```python
# Asegurar que la columna "Precio" sea string antes de limpiarla
df["Precio"] = df["Precio"].astype(str)
```

```python
# Eliminar "S/" y otros caracteres que no sean números o puntos decimales
df["Precio"] = df["Precio"].str.replace(r"[^\d.]", "", regex=True)
```

```python
# Convertir la columna "Precio" a número
df["Precio"] = pd.to_numeric(df["Precio"], errors="coerce")
```

```python
print(df.dtypes)
```

```python
# Eliminar filas con valores NaN en "Precio" (eran valores inválidos)
df = df.dropna(subset=["Precio"])
```

```python
# Eliminar filas con precios negativos o 0
df = df[df["Precio"] > 0]
```

```python
# Recuento de los registros de las variables
df.count()
```

```python
print(df.columns)
```

```python
# Eliminar los espacios
df.columns = df.columns.str.strip()
```

```python
# Todos los nombres de la columna productos que contengan Macbook tendran la marca Apple en la columna Marca
df.loc[df["Producto"].str.contains("MacBook", case=False, na=False), "Marca"] = "Apple"
```

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip()
```

```python
# Todos los nombres de la columna productos que contengan notebook tendran la marca Advance en la columna Marca
mask = (df["Producto"].str.contains("notebook", case=False, na=False)) & (df["Marca"].str.strip() == "No disponible")
```

```python
df.loc[mask, "Marca"] = "Advance"
```

```python
df.head(100)
```

```python
# Convertir todos los datos de la columna Marca a mayuscula
df["Marca"] = df["Marca"].str.upper()
```

```python
df.head(100)
```

```python
data_limpia = df.dropna()
```
