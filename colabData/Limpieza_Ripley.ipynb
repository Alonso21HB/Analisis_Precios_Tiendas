```python
from google.colab import files
import pandas as pd
```

```python
uploaded = files.upload()

df = pd.read_csv("productos_ripley.csv")

df.head(100)
```

```python
#Agregando la columna tienda
df["Tienda"] = "Ripley"
```

```python
df.isnull().sum()
```

```python
# Asegurar que la columna "Precio" sea string antes de limpiarla
df["Precio"] = df["Precio"].astype(str)
```

```python
# Eliminar "S/" y otros caracteres que no sean números o puntos decimales
df["Precio"] = df["Precio"].str.replace(r"[^\d.]", "", regex=True)
```

```python
# Convertir la columna "Precio" a número
df["Precio"] = pd.to_numeric(df["Precio"], errors="coerce")
```

```python
print(df.dtypes)
```

```python
# Eliminar filas con valores NaN en "Precio" (eran valores inválidos)
df = df.dropna(subset=["Precio"])
```

```python
# Eliminar filas con precios negativos o 0
df = df[df["Precio"] > 0]
```

```python
# Recuento de los registros de las variables
df.count()
```

```python
df.columns = df.columns.str.strip()
```

```python
# de la columna Producto solo extrer los datos que contengan laptop, macbook y notebook
df_laptops = df[df["Producto"].str.contains(r"LAPTOP|MACBOOK|NOTEBOOK", regex=True, na=False)]
```

```python
df_laptops.count()
```

```python
data_limpia = df_laptops.dropna()
```

```python
data_limpia.count()
```
